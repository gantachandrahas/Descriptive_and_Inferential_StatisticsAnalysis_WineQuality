```python
## 9.Remove the outliers if any


def find_outliers_IQR(color_wine):
    q1=color_wine.quantile(0.25)
    q3=color_wine.quantile(0.75)
    IQR=q3-q1
    outliers=color_wine[((color_wine<(q1-1.5*IQR)) | (color_wine>(q3+1.5*IQR)))]
    return outliers

outliers = find_outliers_IQR(color_wine[["fixed_acidity","volatile_acidity","citric_acid","residual_sugar","chlorides","free_sulfur_dioxide","total_sulfur_dioxide","density","pH","sulphates","alcohol","quality"]])

print("number of outliers: "+ str(len(outliers)))
print()
print("max outlier value: "+ str(outliers.max()))
print()
print("min outlier value: "+ str(outliers.min()))
```
